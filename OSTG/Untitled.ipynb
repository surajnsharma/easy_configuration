```python
# Simulate the behavior of IPv4 Increment settings

# Sample stream data extracted from JSON
stream_data = {
    "ipv4": {
        "increment_source": True,
        "source_increment_step": "2",
        "source_increment_count": "5",
        "increment_destination": True,
        "destination_increment_step": "3",
        "destination_increment_count": "4"
    }
}

# Simulated UI elements (checkboxes and text fields)
class SimulatedCheckbox:
    def __init__(self):
        self.checked = False

    def setChecked(self, state):
        self.checked = state

    def isChecked(self):
        return self.checked


class SimulatedTextField:
    def __init__(self):
        self.text = ""

    def setText(self, value):
        self.text = value

    def setEnabled(self, state):
        self.enabled = state  # Simulating enabling/disabling

    def getText(self):
        return self.text


# Instantiate simulated UI elements
increment_source_checkbox = SimulatedCheckbox()
source_increment_step = SimulatedTextField()
source_increment_count = SimulatedTextField()
increment_destination_checkbox = SimulatedCheckbox()
destination_increment_step = SimulatedTextField()
destination_increment_count = SimulatedTextField()

# Simulating the logic
increment_source_checkbox.setChecked(stream_data["ipv4"].get("increment_source", False))
source_increment_step.setText(stream_data["ipv4"].get("source_increment_step", "1"))
source_increment_count.setText(stream_data["ipv4"].get("source_increment_count", "1"))
source_increment_step.setEnabled(increment_source_checkbox.isChecked())
source_increment_count.setEnabled(increment_source_checkbox.isChecked())

increment_destination_checkbox.setChecked(stream_data["ipv4"].get("increment_destination", False))
destination_increment_step.setText(stream_data["ipv4"].get("destination_increment_step", "1"))
destination_increment_count.setText(stream_data["ipv4"].get("destination_increment_count", "1"))
destination_increment_step.setEnabled(increment_destination_checkbox.isChecked())
destination_increment_count.setEnabled(increment_destination_checkbox.isChecked())

# Output results
print("Source Increment Checkbox Checked:", increment_source_checkbox.isChecked())
print("Source Increment Step:", source_increment_step.getText())
print("Source Increment Count:", source_increment_count.getText())
print("Source Increment Step Enabled:", getattr(source_increment_step, 'enabled', False))
print("Source Increment Count Enabled:", getattr(source_increment_count, 'enabled', False))
print("Destination Increment Checkbox Checked:", increment_destination_checkbox.isChecked())
print("Destination Increment Step:", destination_increment_step.getText())
print("Destination Increment Count:", destination_increment_count.getText())
print("Destination Increment Step Enabled:", getattr(destination_increment_step, 'enabled', False))
print("Destination Increment Count Enabled:", getattr(destination_increment_count, 'enabled', False))
```
