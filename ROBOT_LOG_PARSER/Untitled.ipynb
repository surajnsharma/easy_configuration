```python
from jnpr.junos import Device
from lxml import etree
import paramiko
from jnpr.junos.exception import ConnectError

config_lines = ["show interface terse", "set system services ssh"]

# Open a connection to the device
dev = Device(host="ny-q5240-13.englab.juniper.net", user="root", passwd="Embe1mpls", port=22)
dev.open()

def is_operational_command(line):
    """ Check if a line is an operational command (e.g., starts with 'show' or 'clear'). """
    return line.strip().lower().startswith(("show", "clear"))

# Filter configuration lines
clean_config_lines = [line for line in config_lines if line.strip() and not line.startswith("#")]
filtered_config_lines = [line for line in clean_config_lines if not is_operational_command(line)]
operational_commands = [line for line in clean_config_lines if is_operational_command(line)]

print("All Commands:", clean_config_lines)
print("Filtered Config Commands:", filtered_config_lines)
print("Operational Commands:", operational_commands)

# Execute operational commands if any
for command in operational_commands:
    try:
        result = dev.rpc.cli(command, format='text')
        print(f"Result of '{command}':", etree.tostring(result, pretty_print=True).decode())
    except ConnectError as e:
        print(f"Error executing '{command}': {e}")

dev.close()
```
