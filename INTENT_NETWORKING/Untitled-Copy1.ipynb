```python
from jnpr.junos import Device
from lxml import etree

# Device details
host = 'svla-q5240-08.englab.juniper.net'
user = 'root'
password = 'Embe1mpls'
port = 22

# Command to execute
xml_command = "show interfaces terse"

try:
    # Open connection to the device
    dev = Device(host=host, user=user, password=password, port=port)
    dev.open()

    # Execute the command and retrieve XML output
    response = dev.rpc.cli(xml_command, format='xml')

    # Convert the XML response to a pretty-printed string
    response_output = etree.tostring(response, pretty_print=True).decode('utf-8')
    print(response_output)

    dev.close()
except Exception as e:
    print(f"Error: {e}")
```
